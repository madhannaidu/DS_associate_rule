```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
movie = pd.read_csv('/content/my_movies.csv')
movie
```

```python
movie.info()
```

```python
movie1=movie.iloc[:,5:]
movie1
```

```python
mov_itemsets=apriori(movie1,min_support=0.1,use_colnames=True)
mov_itemsets
```

```python
rules=association_rules(mov_itemsets,metric='lift',min_threshold=0.7)
rules
```

```python
rules[rules.lift>1]
```

```python
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```
